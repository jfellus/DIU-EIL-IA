# Rseaux de neurones

L'tude des mcanismes physiologiques de l'apprentissage dans le cerveau remonte (au moins)  la fin du XIXme sicle. On cherche alors  montrer comment la dynamique lectrochimique des neurones et la variation des connexions qui les relient peuvent induire un developpement des capacits cognitives.


Source: Samir Chtita

Une des premires thories sur l'apprentissage dans les assembles de neurones est celle de Hebb (1949), qui observe que les liens synaptiques entre un groupe de neurones sont renforcs lorsque ces neurones sont coactivs. Les rseaux hebbiens permettent ainsi de mmoriser l'information en la faisant porter par les poids synaptiques des neurones coactivs.

Les premiers rseaux de neurones formels apparaissent en 1943 avec le modle de Pitt-McCulloch. Ils modlisent de manire simplifie le fonctionnement des neurones biologiques avec des oprateurs mathmatiques lmentaires. Ce modle est ensuite complt par Rosenblatt qui ajoute notamment la rgle d'apprentissage de Widrow-Hoff pour donner le Perceptron.

Un perceptron (simple) est une fonction 

$$ f: (x,w) \mapsto y = \mathcal{H}\Big(\sum w_i \cdot x_i - \theta \Big) $$

* Les $x_i \in \mathbb{R}$ sont les entres, formant le vecteur d'entre $x \in \mathbb{R}^n$  $n$ dimensions
* $y = f(x_1,\dots,x_n)$ est la sortie
* Les $w_i  \in \mathbb{R}$ sont les poids synaptiques forment un vecteur $w \in \mathbb{R}^n$  $n$ dimensions.
* $\theta  \in \mathbb{R}$ est un seuil
* $\mathcal{H}$ est la fonction de Heaviside, telle que $\mathcal{H}(x)=1$ si $x\geq 0$ et $0$ sinon

Formellement, un perceptron est donc un classificateur binaire. C'est galement un discriminateur linaire: la frontire de dcision est un hyperplan d'quation $\sum x_i.w_i = \theta$

Les poids $w_i$ sont les paramtres ajustables du modle. 

Les paramtres sont mis  jour par la rgle d'apprentissage de Widrow-Hoff :

$$ w_i(t+1) = w_i(t) + \lambda\Big(y^*(t)-f(x_i(t)\Big)x_i(t) $$

* A chaque itration $t$, on fournit un vecteur $x(t)$ et un rel $y^*(t)$ 
* $y^*(t)$ est la sortie dsire associe  $x(t)$
* $\lambda$ est le pas d'apprentissage ou *learning rate*.

Le perceptron est le modle le plus simple de neurone. Tous les modles de neurones formels incorporent :
* Le produit scalaire entre les poids et les entres, nomm "potentiel"
* Une fonction d'activation, gnralement non-linaire (sigmode, ReLU, seuil, gaussienne, *etc*)


```python
# Exemple : Perceptron simple

import numpy as np
import matplotlib.pyplot as plt

def perceptron(w,x,th):
    return sum(w*x)-th >= 0

w = np.array([-.5,1])
th = 0.3

x = np.random.rand(1000,2)

plt.scatter(*x.T, c=['b' if perceptron(w,x,th) else 'r' for x in x])
plt.plot([0,1],[th/w[1],(th-w[0])/w[1]], linewidth=4, c='g')
```

```python
# Un perceptron plutt mauvais

w = np.array([-.5,1])
th = 0.3

x = np.random.rand(1000,2)
y = x[:,0]>0.5

plt.scatter(*x.T, c=['yellow' if y else 'purple' for y in y])
plt.plot([0,1],[th/w[1],(th-w[0])/w[1]], linewidth=4, c='g')
```

# Descente de gradient 

Notons $MSE = \frac{1}{2N}\sum_i \Big[y_i - f(x_i,w_i)\Big]^2$ l'erreur quadratique moyenne (divise par 2) de notre perceptron o $(x_i,y_i)$ designe le i-me couple d'apprentissage (entre, sortie dsire).

Remarquons que la MSE est drivable par rapport aux $w_i$ ssi $f$ est drivable par rapport aux $w_i$.
Lorsque c'est le cas, le gradient $\nabla MSE = (\frac{\partial MSE}{\partial w_i})_i$ nous donne la direction dans laquelle l'erreur crot. Si l'on veut diminuer l'erreur, il suffit donc de suivre la direction oppose : c'est la descente de gradient

Notre perceptron $f$ n'est pas directement drivable  cause de la fonction $\mathcal{H}$ (non-derivable en 0). L'ide de Widrow et Hoff est de mettre  jour les poids en ne considrant pas la fonction d'activation. L'algorithme Adaline de Widrow et Hoff effectue alors une descente de gradient directement sur le potentiel du neurone :

$$E = \frac{1}{2N} \sum_i \Big[y_i - <x_i,w> + \theta\Big]^2$$

La drive partielle de cette erreur par rapport  chaque poids $w_j$ se calcule facilement :

$$\frac{\partial E}{\partial w_j} = -\frac{1}{N} \sum_i \Big[y_i - <x_i,w> + \theta\Big]x_i $$

On suit alors la procdure d'optimisation itrative suivante :
* **Evaluation** : on calcule le potentiel du neurone pour chaque entre du jeu de donnes
* **Calcul du gradient** : on calcule les derives partielles par rapport  chaque poids : 
$$\nabla E = (\frac{\partial E}{\partial w_i})_i$$
* **Descente** : on dplace les poids dans la direction ngative du gradient, c'est  dire :
$$w(t+1) = w(t) - \alpha \nabla E$$


Lorsque la mesure d'erreur et le modle sont convexes, la descente de gradient converge vers le minimum global.
Les fonctions non-convexes ont plusieurs minima locaux (des points o le gradient est nul), dans ce cas la descente de gradient ne garantit pas de trouver celui qui est globalement optimal.


```python
# Illustration : descente de gradient pour minimiser une fonction quadratique
from IPython.display import clear_output
import time

def f(x):
    return x**2

def derivee(x):
    return 2*x

def show_f():
    plt.xlim([-10,10])
    xf = np.arange(-10,10,.1)
    yf = f(xf)
    plt.plot(xf, yf, '-')

x = -10
for i in range(40):
    y = f(x)
    grad = derivee(x)
   
    clear_output(True)
    show_f()
    plt.arrow(x,y,-0.05*grad,-0.05*grad**2, width=0.5)
    plt.show()
    time.sleep(1)

    x = x - 0.02*grad

    
```

```python
# Exemple : Entraner un Perceptron avec la descente de gradient

import numpy as np
from IPython.display import clear_output
import matplotlib.pyplot as plt
import time


def H(x):
    return int(x>=0)

def perceptron(w,x,th):
    return H(sum(w*x)-th)

def grad_w(w,th, X,Y):
    return -1/(2*len(Y)) * sum([(y - perceptron(w,x,th))*x for x,y in zip(X,Y)])
    
def grad_th(w,th, X,Y):
    return 1/(2*len(Y)) * sum([(y - perceptron(w,x,th)) for x,y in zip(X,Y)])

w = np.array([-.5,1])
th = 0

X = np.random.rand(1000,2)
Y = (X[:,0]+X[:,1])>0.7


def show():
    clear_output(True)
    plt.xlim([0,1])
    plt.ylim([0,1])
    plt.scatter(*X.T, c=['b' if y else 'r' for y in Y],marker='o')
    plt.scatter(*X.T, c=['b' if perceptron(w,x,th) else 'r' for x in X],marker='+')
    plt.plot([0,1],[th/w[1],(th-w[0])/w[1]], linewidth=4, c='g')
    plt.show()
    
for i in range(500):
    _w = w
    w -= 1*grad_w(_w,th, X,Y)
    th -= 1*grad_th(_w,th, X,Y)
    show()
    time.sleep(.001)
```

# Regression logistique

Observons qu'avec une activation Heaviside, notre modle n'est pas fortement-convexe. Autrement dit il existe des "plateaux" dans la fonction de cot. 
Dans le cas du Perceptron, ce plateau se manifeste justement pour E=0.

Cela signifie qu'il peut exister une infinit de Perceptron optimaux pour un problme donn.

Illustration :

```python
# Plusieurs Perceptrons optimaux en entranement

w = np.array([-.8,0.01])
th = -0.3

x = np.array([[0.1,0.1],[0.1,0.5], [0.8,0.2],[0.8,0.3]])
y = x[:,0]>0.5

plt.xlim([0,1])
plt.ylim([0,1])
plt.scatter(*x.T, c=['b' if y else 'r' for y in y])
plt.plot([0,1],[th/w[1],(th-w[0])/w[1]], linewidth=4, c='g')

w = np.array([-.4,0.01])
th = -0.3
plt.plot([0,1],[th/w[1],(th-w[0])/w[1]], linewidth=4, c='g')
```
